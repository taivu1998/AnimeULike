```
import urllib
import bs4
import re
```

```
url="http://myanimelist.net/"
source = urllib.request.urlopen(url).read()
soup = bs4.BeautifulSoup(source, 'lxml')
```

```
matches_total = []
for entry_no in range(0, 1000, 50):
    url='https://myanimelist.net/topanime.php?limit=0'
    source = urllib.request.urlopen(url).read()
    soup = bs4.BeautifulSoup(source, 'lxml')
    lis = soup.find_all('h3', class_="hoverinfo_trigger fl-l fs14 fw-b anime_ranking_h3")
    pattern = re.compile(r"/anime/([^\"]+)")
    for item in lis:
        str_ = str(item)
        matches = pattern.finditer(str_)
        matches_total = matches_total + [i.group(0) for i in matches]
```

```
print(matches_total)
```

```
complete = ["https://myanimelist.net" + match for match in matches_total]
```

```
print(complete)
```

```
file1 = open("top_links.csv","w") 
file1.write("MAL Anime ID, MAL link\n")

for string in complete:
    # lazy extraction (use re for more robustness)
    num = string[30:35]
    index = num.find('/')
    if(index != -1):
        num = num[0:index]
    file1.write("%s, %s\n"%(num, string))

file1.close() 
```
